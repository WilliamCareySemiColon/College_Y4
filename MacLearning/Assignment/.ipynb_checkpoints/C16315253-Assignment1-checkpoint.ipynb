```python
#libraries to import
import statistics as stat
import pandas as pd

#reading and parsing the potential columns head from the text file
columnsName = open(".//Data//feature_names.txt")
text_reader = columnsName.read()

#setting up the csv headers for future use

CSVColumnHeaders = ["FeatureName"] + text_reader.split("\n")

#parsing the csv file now, first convert from txt file to csv file as necessary

dataTxtFile = pd.read_csv(".//Data//dataset.txt")

#dataTxtFile.head()

csvDir = ".//Data//dataset.csv"

dataTxtFile.to_csv(csvDir)

df = pd.read_csv(csvDir, header = None)

#setting the column headers for the report

df.columns = ["FeatureName"] + text_reader.split("\n")

df.head()
```

```python
ContIndexes = ["age","fnlwgt","education-num","capital-gain","capital-loss","hours-per-week"]
for key in df.columns:
    if(key in ContIndexes):
        Min = df[key].min()
        Max = df[key].max()
        #all are .xxxx
        Median = df[key].median()
        Mean = df[key].mean()
        FirstQrt = df[key].quantile(.25)
        ThirdQrt = df[key].quantile(.75)
        Std = df[key].std()
        print(Min)
        print(Max)
        #format
        print(Mean)
        print(Median)
        print(FirstQrt)
        print(ThirdQrt)
        #format
        print(Std)
        print("\n")
```
