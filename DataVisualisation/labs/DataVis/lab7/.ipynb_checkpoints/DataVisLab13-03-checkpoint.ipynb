```r
#install.packages("sqldf")
library(dplyr)
library(tidyverse)
library(ggplot2)
library(lubridate)
library(sqldf)
```

```r
#reading the general elections file
titanicDF = read.csv("titanic.csv",header=TRUE, sep=",")

head(titanicDF)
```

```r
#All the choosen headers have small cardianlity rate show the decison to set them as ratio bar charts made sense

ggplot(data=titanicDF,aes(titanicDF$Sex)) + geom_bar()

ggplot(data=titanicDF,aes(titanicDF$Pclass)) + geom_bar()

ggplot(data=titanicDF,aes(titanicDF$Embarked)) + geom_bar()

ggplot(data=titanicDF,aes(titanicDF$Survived)) + geom_bar()
```
