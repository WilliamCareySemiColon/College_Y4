```r
#This speech was originally delivered as the commencement address at Harvard University on June 5, 2008.  
#The full title of this talk is “The Fringe Benefits of Failure, and the Importance of Imagination”.
#Delivered by JK Rowling
#First published by the harvard gazetee
#url:https://news.harvard.edu/gazette/story/2008/06/text-of-j-k-rowling-speech/

#install.packages("tm")  # for text mining
#install.packages("SnowballC") # for text stemming
#install.packages("wordcloud") # word-cloud generator 
#install.packages("RColorBrewer") # color palettes

library("tm")
library("SnowballC")
library("wordcloud")
library("RColorBrewer")
```

```r
filepath = "data/JKRowling.txt"
text <- readLines(filepath)
docs <- Corpus(VectorSource(text))
inspect(docs)
```

```r
toEmptySpace <- content_transformer(function (x , pattern ) gsub(pattern, "", x))

#removing the unnecessary characters
docs <- tm_map(docs, toEmptySpace, "â")
docs <- tm_map(docs, toEmptySpace, "\200")
docs <- tm_map(docs, toEmptySpace, "\230")
docs <- tm_map(docs, toEmptySpace, "\231")
docs <- tm_map(docs, content_transformer(tolower))
# Convert the text to lower case
docs <- tm_map(docs, content_transformer(tolower))
# Remove numbers
docs <- tm_map(docs, removeNumbers)
# Remove english common stopwords
docs <- tm_map(docs, removeWords, stopwords("english"))
# Remove punctuations
docs <- tm_map(docs, removePunctuation)
# Eliminate extra white spaces
docs <- tm_map(docs, stripWhitespace)
inspect(docs)
```

```r
dtm <- TermDocumentMatrix(docs)

m <- as.matrix(dtm)
v <- sort(rowSums(m),decreasing=TRUE)
d <- data.frame(word = names(v),freq=v)
head(d, 10)
```

```r
set.seed(1234)# This makes the cloud reproducable.

wordcloud(words = d$word, freq = d$freq, min.freq = 1,
          max.words=200, random.order=FALSE, rot.per=0.35, 
          colors=brewer.pal(8, "Dark2"))
```

```r
barplot(d[1:10,]$freq, las = 2, names.arg = d[1:10,]$word,
        col ="red", main ="Most frequent words",
        ylab = "Word Frequencies")
```

```r
GetCloudAndFrequencyPlot <- function(textfile){
    dtm <- TermDocumentMatrix(textfile)

    m <- as.matrix(dtm)
    v <- sort(rowSums(m),decreasing=TRUE)
    d <- data.frame(word = names(v),freq=v)
    print(head(d, 10))
    set.seed(1234)

    print(wordcloud(words = d$word, freq = d$freq, min.freq = 1,
          max.words=200, random.order=FALSE, rot.per=0.35, 
          colors=brewer.pal(8, "Dark2")))
}

GetCloudAndFrequencyPlot(docs)
```

```r
#https://www.google.com/search?q=thunder+song+lyrics&oq=thunder+song+lyrics&aqs=chrome..69i57j0l7.9568j0j9&sourceid=chrome&ie=UTF-8
#Source: LyricFind
#Songwriters: Alexander Junior Grant / Benjamin Arthur McKee / Daniel Coulter Reynolds / Daniel James Platzman / Daniel Wayne Sermon / Jayson M. DeZuzio
#Thunder lyrics © Universal Music Publishing Group

songfilepath = "data/thunder.txt"
songtext <- readLines(songfilepath)
songDoc <- Corpus(VectorSource(songtext))
# Convert the text to lower case
songDoc <- tm_map(songDoc, content_transformer(tolower))
# Remove numbers
songDoc <- tm_map(songDoc, removeNumbers)
# Remove english common stopwords
songDoc <- tm_map(songDoc, removeWords, stopwords("english"))
# Remove punctuations
songDoc <- tm_map(songDoc, removePunctuation)
# Eliminate extra white spaces
songDoc <- tm_map(songDoc, stripWhitespace)
#inspect(songDoc)

GetCloudAndFrequencyPlot(songDoc)
```
